```python
import plotly.express as px
import numpy as np
import pandas as pd
   

```

```python
from scipy.stats import skewnorm


data = []
for skew in range(-10, 10 + 1):
    x = skewnorm.rvs(skew, size=100000)
    counts, bins = np.histogram(x, bins=np.arange(-5, 5, 0.1))
    bins = bins[:-1]
    
    r = pd.DataFrame(dict(counts=counts, bins=bins))
    r['skew'] = skew
    data.append(r)
    
data = pd.concat(data, axis=0)

fig = px.scatter(data, x='bins', y='counts', animation_frame='skew', animation_group='bins', range_x=[-5, 5], range_y=[0, data.counts.max()])

with open('tmp.html', 'w') as f:
    f.write(fig.to_html())
fig
```
